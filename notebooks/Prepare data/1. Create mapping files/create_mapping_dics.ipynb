```python
class_str_to_class_int = dict()
count = 0
for k in class_to_subclass.keys():
    class_str_to_class_int[k] = count
    count += 1
```

```python
subclass_to_class_str = dict()
subclass_to_class_int = dict()
for k, v in class_to_subclass.items():
    for subclass in v:
        subclass_to_class_str[subclass] = k
        subclass_to_class_int[subclass] = class_str_to_class_int[k]
```
